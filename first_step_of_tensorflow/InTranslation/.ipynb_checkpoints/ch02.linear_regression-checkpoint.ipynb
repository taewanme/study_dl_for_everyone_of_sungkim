## 학습 데이터

```python
import numpy as np

num_point = 1000
vector_set = []

for i in range(num_point):
    x1 = np.random.normal(0.0, 0.55)
    y1 = x1 * 0.1 + 0.3 + np.random.normal(0.0, 0.03)
    vector_set.append([x1, y1])
    
x_data = [v[0] for v in vector_set]
y_data = [v[1] for v in vector_set]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x_data, y_data, 'ro')
plt.legend()
plt.show()
```

## 2.2 비용함수와 경사 하강법 알고리즘

- cost function, loss function
  - mean square error
  - 평균 제곱 오차

```python
import tensorflow as tf
```

```python
W = tf.Variable(tf.random_normal([1], -1.0, 1.0), name='weight')
b = tf.Variable(tf.zeros([1]))
y = W * x_data + b
```

```python
cost = tf.reduce_mean(tf.square(y-y_data))
```

```python
train = tf.train.GradientDescentOptimizer(0.2).minimize(cost)
```

## 2.3 알고리즘 실행

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for step in range(200):
        a, loss = sess.run([train,cost])
        if step % 40 == 0:
            print(step, loss, sess.run(W), sess.run(b))
            plt.plot(x_data, y_data, "ro")
            plt.plot(x_data, sess.run(W)*x_data+sess.run(b))
            plt.show()
```

- learning rate: 각 반복에서 이동하는 크기

## 전체 소스

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

num_points = 1000
vector_set = []

for i in range(num_points):
    x1 = np.random.normal(0.0, 0.55)
    y1 = x1*0.1 + 0.3 + np.random.normal(0.0, 0.03)
    vector_set.append([x1, y1])
```

```python
x_data = [v[0] for v in vector_set]
y_data = [v[1] for v in vector_set]
```

```python
W = tf.Variable(tf.random_uniform([1], -1.0, 1.0))
b = tf.Variable(tf.zeros([1]))
y = W*x_data + b
```

```python
cost = tf.reduce_mean(tf.square(y-y_data))
```

```python
train = tf.train.GradientDescentOptimizer(0.1).minimize(cost)
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    for step in range(200):
        _, loss = sess.run([train,cost])
        if step % 40 == 0 :
            print("cost: ", loss, " weight: ", sess.run(W), " bias: ", sess.run(b))
            plt.plot(x_data, y_data, "ro")
            plt.plot(x_data, x_data*sess.run(W)+sess.run(b))
            plt.xlabel('x')
            plt.ylabel('y')
            plt.xlim(-2, 2)
            plt.ylim(0., 0.6)
            plt.legend()
            plt.show()
            
    
```
